```python
import numpy as np
import matplotlib.pylab as plt
from pylab import rcParams
rcParams['figure.figsize'] = 15, 8
```

```python
#Creating data for polynomial regression
x = np.linspace(1,20,num=15)
y = np.log(x)
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
#Ploting data
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Creating Polynominal Features
from sklearn.preprocessing import PolynomialFeatures 
poly = PolynomialFeatures(degree=2)
x_train = poly.fit_transform(x)
x_train.shape
```

```python
#Importing and training our implementation of Regularized Linear Regression 
from reg_linear_regression import LinearRegression
clf_lasso = LinearRegression(reg = "l1",lmbda=.1)
theta_lasso = clf_lasso.fit(x_train,y)
clf_ridge = LinearRegression(reg = "l2",lmbda=8)
theta_ridge = clf_ridge.fit(x_train,y)
clf_linear = LinearRegression(reg = None)
theta_linear = clf_linear.fit(x,y)
```

```python
y_pred_lasso = clf_lasso.predict(x_train,theta_lasso)
y_pred_ridge = clf_ridge.predict(x_train,theta_ridge)
y_pred_linear = clf_linear.predict(x,theta_linear)
```

```python
#Ploting fitted curves for training
plt.scatter(x,y,marker="^")
plt.plot(x,y_pred_lasso,color="red",label="Lasso",linewidth=2)
plt.plot(x,y_pred_ridge,color="green",label="Ridge",linewidth=2)
plt.plot(x,y_pred_linear,color="magenta",label="Linear",linewidth=2)
plt.title("Lasso Ridge & Linear Regression")
plt.legend()
plt.show()
```

```python
#importing sklearn's linear Models
from sklearn.linear_model import Ridge,Lasso,LinearRegression
clf_lasso = Lasso(alpha=.2).fit(x_train,y)
y_pred_lasso=clf_lasso.predict(x_train)
clf_ridge = Ridge(alpha=5).fit(x_train,y)
y_pred_ridge=clf_ridge.predict(x_train)
clf_linear = LinearRegression().fit(x,y)
y_pred_linear=clf_linear.predict(x)
```

```python
#Ploting curves fitted by Sklearn 
plt.scatter(x,y,marker="^")
plt.plot(x,y_pred_lasso,color="red",label="Lasso",linewidth=2)
plt.plot(x,y_pred_ridge,color="green",label="Ridge",linewidth=2)
plt.plot(x,y_pred_linear,color="magenta",label="Linear",linewidth=2)
plt.title("Lasso Ridge & Linear Regression(Sklearn)",)
plt.legend()
plt.show()
```
